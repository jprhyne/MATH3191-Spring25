```python
import numpy as np
```

```python
maxSearch = 4
numMats = (maxSearch+1)**4
numTriMats = (maxSearch+1)**3
AMatVec = [np.zeros((2,2)) for i in range(numMats)]
BMatVec = [np.zeros((2,2)) for i in range(numTriMats)]
CMatVec = [np.zeros((2,2)) for i in range(numMats)]
currentIndex = 0
for a in range(maxSearch+1):
    for b in range(maxSearch+1):
        for c in range(maxSearch+1):
            for d in range(maxSearch+1):
                AMatVec[currentIndex] = np.array([[a,b],[c,d]])
                CMatVec[currentIndex] = np.array([[a,b],[c,d]])
                currentIndex = currentIndex + 1
```

```python
currentIndex = 0
for a in range(maxSearch+1):
    for b in range(maxSearch+1):
        for c in range(maxSearch+1):
            BMatVec[currentIndex] = np.array([[a,b],[0,c]])
            currentIndex = currentIndex + 1
```

```python
for A in AMatVec:
    if (A[0,0]==0.0 or A[0,1]==0.0 or A[1,0]==0.0 or A[1,1]==0.0):
        continue
    for B in BMatVec:
        if (B[0,0]==0.0 or B[0,1]==0.0 or B[1,1]==0.0 or B[0,0]==B[1,1]):
            continue
        for C in CMatVec:
            if (C[0,0]*C[1,1]-C[0,1]*C[1,0] == 0 or (C[0,0]==C[1,1] and C[0,1]==0.0)):
                continue
            LHS = A@C
            RHS = C@B
            if np.linalg.norm(LHS-RHS) == 0.0:
                # Check if B has integer eigenvalues and eigenvectors
                # we know that the eigenvalues are integer but its eigenvectors are only
                # integral if B[0,1]/(B[0,0]-B[1,1]) is an integer
                testVal = B[0,1]/(B[0,0]-B[1,1])
                if int(testVal) == testVal:
                
                    print("----------------------------")
                    print("A="+str(A))
                    print("B="+str(B))
                    print("C="+str(C))
                    print("----------------------------")
```
